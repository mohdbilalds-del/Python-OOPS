**Q1- What is Object-Oriented Programming (OOP)?**
- **Object-Oriented Programming (OOP)** in Python is a programming paradigm that organizes software design around "objects" rather than functions and logic. **Q2- What is a class in OOP?**

**Q3- What is an object in OOP?**

**Q4- What is the difference between abstraction and encapsulation?**
- Abstraction hides complexity by showing only essential features (the "what"), while Encapsulation bundles data and methods into one unit and restricts direct access (the "how"), with abstraction often implemented using encapsulated objects for a cleaner interface, like driving a car without knowing engine details. Abstraction focuses on design/interface, using abstract classes/interfaces; encapsulation focuses on implementation/security, using access modifiers (private/public).

**Q9- What is a constructor in Python?**

**Q10- What are class and static methods in Python?**
- In Python, both class methods and static methods are defined within a class but differ in their binding and access to class or instance data.

  **Class Methods:**
  - **Binding:** Class methods are bound to the class, not an instance of the class. They receive the class itself as their first argument, conventionally named cls.
  - **Decorator:** They are defined using the ***@classmethod*** decorator.
  - **Access:** Class methods can access and modify class-level attributes, but they cannot directly access or modify instance-level attributes.
  - **Use Cases:** Common uses include:
    - **Factory methods:** Providing alternative ways to construct instances of the class.
    - **Class-level operations:** Methods that operate on or require knowledge of the class's state or attributes.

   **Static Methods:**
  -  **Binding:** Static methods are not bound to either the class or any instance of the class. They do not receive self or cls as their first argument.
  -  **Decorator:** They are defined using the ***@staticmethod*** decorator.
  -  **Access:** Static methods cannot access or modify class-level or instance-level attributes directly. They are essentially regular functions logically grouped within a class.
  -  **Use Cases:** Primarily used for utility or helper functions that have a logical connection to the class but do not require access to the class's state or instance-specific data.

**Q11- What is method overloading in Python?**

The appropriate method version to execute is determined by the arguments provided during the function call.

**Q12- What is method overriding in OOP?**

It allows customizing inherited behavior, so a Dog class can move() by running, while a Fish class (also inheriting from Animal) can move() by swimming, overriding the general Animal's move() method.

**Q17- What is the difference between a class variable and an instance variable?**
- The key difference is that a class variable is shared by all instances of a class, while an instance variable is unique to each individual instance (object)

    **Class Variables**
  -  **Ownership:** Owned by the class itself, not a specific object.
  -  **Sharing:** There is only one copy of a class variable, and all instances share that single copy.
  -  **Declaration:** In most languages like Java, they are declared using the static keyword. In Python, they are defined directly within the class but outside any methods.
  -  **Access:** Can be accessed using either the class name (e.g., **ClassName.variable_name**) or an object reference.
  -  **Lifetime:** Created when the program starts (or when the class is first loaded) and destroyed when the program terminates.
  -  **Use Cases:** Ideal for storing data that is consistent across all instances, such as a constant value, a shared configuration, or a count of how many objects have been created.

    **Instance Variables**
  -  **Ownership:** Owned by a specific instance (object) of the class.
  -  **Sharing:** Each object has its own personal copy of the instance variable, independent of other objects.
  -  **Declaration:** Declared without the static keyword in languages like Java. In Python, they are typically defined within the constructor method (e.g., **_ _ init _ _**) using the self keyword.
  -  **Access:** Accessed using the object reference (e.g., **object_name.variable_name**).
  -  **Lifetime:** Created when an object is instantiated (using new in Java or by calling the class in Python) and destroyed when the object is destroyed.
  -  **Use Cases:** Ideal for storing data that defines the unique state or properties of each object, such as a person's name or age.

**Q19- Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python.**
- In Python, __ str __ and __ repr __ are "dunder" (double-underscore) methods that provide string representations of objects. They serve different primary purposes: __str__ is designed for readability by humans, while __repr__ is intended to be unambiguous and ideally machine-readable.

   **__ str __ (String Representation)**
   
   The __ str __ method is called by the built-in str() function and generally by print() [1, 3]. Its goal is to return a concise, user-friendly, or "pretty" string that is easy for a person to read and understand.

   **__ repr __ (Representation)**
   
   The __ repr __ method is called by the built-in repr() function and is typically what the Python interpreter uses when an object is inspected directly (e.g., in a console without using print). The returned string should be a canonical representation of the object, often looking like the code needed to recreate it.

**Q21- What is the significance of the __del__ method in Python?**
- The __del__ method, also known as the destructor, is a special method in Python classes that is called when an instance of a class is about to be destroyed or finalized.
Its primary significance lies in performing cleanup actions before the object is removed from memory.

    **Key aspects of its significance:**

    **Resource Management:** It is used to release external resources that an object holds, such as closing open files, network connections, or database connections, which are not automatically handled by Python's garbage collector [1].
    
    **Finalization:** It allows a developer to specify actions that must occur when an object is no longer in use, ensuring that the program is left in a clean state [1].
    
    **Contrast with __ init __:** While the __ init __ method (constructor) sets up resources when an object is created, the __ del __ method tears them down when the object is destroyed.

**Q22- What is the difference between @staticmethod and @classmethod in Python?**
- The core difference between **@staticmethod** and **@classmethod** in Python lies in how they access class or instance data:

    **@staticmethod:** Does not receive the class or the instance as its first argument. It's a regular function logically grouped within the class and cannot access any class or instance-specific data.

    **@classmethod:** Receives the class itself (conventionally named cls) as its first argument. It can access and modify class-level attributes, and is commonly used for alternative constructors or methods that modify the class state.

**Q23- How does polymorphism work in Python with inheritance?**
- In Python, polymorphism with inheritance is achieved primarily through method overriding, where a subclass redefines a method that is already present in its parent class. This allows objects of different classes that share a common base class to respond to the same method call in a way specific to their own type.

  **How it Works**

  The core idea is the "is-a" relationship (e.g., a Dog is an Animal, a Car is a Vehicle).

  - **Inheritance** establishes a hierarchy, allowing child classes to inherit methods and attributes from a parent class.

  - **Method Overriding** allows the child class to provide a specific implementation of an inherited method. Python's dynamic typing system determines which method to call at runtime based on the actual object type, not the variable type referring to it. All methods in Python are effectively "virtual" by default, meaning they can be overridden without special syntax.

**Q24- What is method chaining in Python OOP?**
- Method chaining in Python OOP is a technique for calling multiple methods on the same object in a single, continuous line of code. This is achieved by having each method in the chain return the object instance itself, typically using return self.

  **How it Works**

  The core principle is that the return value of one method becomes the object on which the next method in the chain is called. This eliminates the need for intermediate variables to store the object at each step of a series of operations.

  In a Python class, you can enable method chaining by ensuring your methods explicitly return self, the current instance of the class.

```python
#Q2- Write a program to create an abstract class Shape with a method area().
#Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Rectangle class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

circle = Circle(5)
print("Circle area:", circle.area())

rectangle = Rectangle(4, 6)
print("Rectangle area:", rectangle.area())
```

```python
#Q3- Implement a multi-level inheritance scenario where a class Vehicle has an attribute type.
#Derive a class Car and further derive a class ElectricCar that adds a battery attribute

class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

    def show_type(self):
        print(f"Vehicle type: {self.type}")

class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)   # Call Vehicle constructor
        self.brand = brand

    def show_details(self):
        print(f"Car brand: {self.brand}, Type: {self.type}")

# Further derived class from Car
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)   # Call Car constructor
        self.battery = battery_capacity

    def show_info(self):
        print(f"Electric Car brand: {self.brand}, Type: {self.type}, Battery: {self.battery} kWh")

vehicle = Vehicle("General Vehicle")
vehicle.show_type()   # Output: Vehicle type: General Vehicle

car = Car("Sedan", "Toyota")
car.show_details()    # Output: Car brand: Toyota, Type: Sedan

ev = ElectricCar("Hatchback", "Tesla", 75)
ev.show_info()
```

```python
#Q5- Write a program to demonstrate encapsulation by creating a class BankAccount
#with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, initial_balance=0):
        # Private attribute
        self.__balance = initial_balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Insufficient balance or invalid amount.")

    # Method to check balance
    def check_balance(self):
        print(f"Current Balance: {self.__balance}")


account = BankAccount(100)
account.check_balance()

account.deposit(50)
account.check_balance()

account.withdraw(70)
account.check_balance()

account.withdraw(200)
```

```python
#Q7- Create a class MathOperations with a class method add_numbers() to add two numbers
#and a static method subtract_numbers() to subtract two numbers.

class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b


# Calling class method
result_add = MathOperations.add_numbers(10, 5)
print("Addition:", result_add)

# Calling static method
result_sub = MathOperations.subtract_numbers(10, 5)
print("Subtraction:", result_sub)
```

```python
#Q9- Write a class Fraction with attributes numerator and denominator.
#Override the str method to display the fraction as "numerator/denominator".

class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"


f1 = Fraction(3, 4)
print(f1)

f2 = Fraction(5, 3)
print(f2)
```

```python
#Q10- Demonstrate operator overloading by creating a class Vector and
#overriding the add method to add two vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"


v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2
print("Resultant Vector:", v3)
```

```python
#Q12- Implement a class Student with attributes name and grades.
#Create a method average_grade() to compute the average of the grades.

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # Expecting a list of numbers

    def average_grade(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)


s1 = Student("Bilal", [95, 90, 98])
print(f"{s1.name}'s average grade: {s1.average_grade()}")

s2 = Student("Mohd", [92, 88, 95, 91])
print(f"{s2.name}'s average grade: {s2.average_grade()}")
```

```python
#Q14- Create a class Employee with a method calculate_salary() that computes the salary based on
#hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)  # Call Employee constructor
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus


emp = Employee("Bilal", 40, 20)
print(f"{emp.name}'s Salary: ${emp.calculate_salary()}")

mgr = Manager("Mohd", 40, 30, 500)  # 40 hours, $30/hour + $500 bonus
print(f"{mgr.name}'s Salary: ${mgr.calculate_salary()}")
```

```python
#Q15- Create a class Product with attributes name, price, and quantity.
#Implement a method total_price() that calculates the total price of the product.

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity


p1 = Product("Laptop", 50000, 2)
print(f"Total price of {p1.name}: ₹{p1.total_price()}")

p2 = Product("Phone", 20000, 3)
print(f"Total price of {p2.name}: ₹{p2.total_price()}")
```

```python
#Q16- Create a class Animal with an abstract method sound().
#Create two derived classes Cow and Sheep that implement the sound() method.

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Cow says: Moo!")

class Sheep(Animal):
    def sound(self):
        print("Sheep says: Baa!")

animals = [Cow(), Sheep()]

for animal in animals:
    animal.sound()
```

```python
#Q17- Create a class Book with attributes title, author, and year_published.
#Add a method get_book_info() that returns a formatted string with the book's details.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"


b1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925)
print(b1.get_book_info())

b2 = Book("To Kill a Mockingbird", "Harper Lee", 1960)
print(b2.get_book_info())
```
